```python
import pandas as pd
```

```python
df = pd.read_csv('cardiovascular_diseases_dv3.csv', sep=';', quoting=3)
```

Drop columns

```python
df.drop('AP_LOW', inplace=True, axis=1)
df.drop('AP_HIGH',inplace=True, axis=1)
df
```

Compute the BMI, BMI = WEIGHT/HEIGHT^2

```python
df['BMI'] = df['WEIGHT']/(df['HEIGHT']/100)**2
```

```python
df.drop('HEIGHT', inplace=True, axis=1)
df.drop('WEIGHT', inplace=True, axis=1)
df
```

Order the columns

```python
col_names = ['AGE', 'BMI', 'GENDER', 'CHOLESTEROL', 'GLUCOSE', 'SMOKE', 'ALCOHOL', 'PHYSICAL_ACTIVITY', 'CARDIO_DISEASE']
df = df.reindex(columns=col_names)
df
```

Categorize the BMI column: 

Underweight: <18.5 ; Normalweight: 18.5 - 25 ; Overweight: 25.0 - 30 ; Obese: >30

```python
category = pd.cut(df.BMI,bins=[0,18.5,25,30,1000],labels=['Under','Normal','Over','Obese'])
df.insert(5,'BMI Group',category)
df
```

```python
df.drop('BMI', inplace=True, axis=1)
df = df.rename(columns={'BMI Group': 'BMI'})
df
```

Order the columns

```python
col_names = ['AGE', 'BMI', 'GENDER', 'CHOLESTEROL', 'GLUCOSE', 'SMOKE', 'ALCOHOL', 'PHYSICAL_ACTIVITY', 'CARDIO_DISEASE']
df = df.reindex(columns=col_names)
df
```

From continuous to category variables

```python
category = pd.cut(df.AGE,bins=[0,40,50,60,1000],labels=['30-39','40-49','50-59','60-65'])
df.insert(5,'AGE Group',category)
df.drop('AGE', inplace=True, axis=1)
df = df.rename(columns={'AGE Group': 'AGE'})
col_names = ['AGE', 'BMI', 'GENDER', 'CHOLESTEROL', 'GLUCOSE', 'SMOKE', 'ALCOHOL', 'PHYSICAL_ACTIVITY', 'CARDIO_DISEASE']
df = df.reindex(columns=col_names)
df
```

Recode columns values

```python
df['GENDER'] = df['GENDER'].replace([1,2],['Female','Male'])
df['CHOLESTEROL'] = df['CHOLESTEROL'].replace([1,2,3],['Normal','Moderate','High'])
df['GLUCOSE'] = df['GLUCOSE'].replace([1,2,3],['Normal','Moderate','High'])
df = df.replace([0],'No')
df = df.replace([1],'Yes')
df
```

Convert type to category

```python
for col in ['GENDER', 'CHOLESTEROL', 'GLUCOSE', 'SMOKE', 'ALCOHOL', 'PHYSICAL_ACTIVITY', 'CARDIO_DISEASE']:
    df[col] = df[col].astype('category')
```

Rename columns to lowercase

```python
df = df.rename(columns={'AGE': 'Age', 'GENDER': 'Gender','CHOLESTEROL': 'Cholesterol', 'GLUCOSE': 'Glucose', 'SMOKE':'Smoke' , 'ALCOHOL':'Alcohol' , 'PHYSICAL_ACTIVITY':'PhysicalActivity' , 'CARDIO_DISEASE':'CardioDisease'})
df
```

Export CSV

```python
df.to_csv(path_or_buf='cardiovascular_disease_clean.csv', sep=',', index=False)
```
